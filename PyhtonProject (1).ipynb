```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("C:/Users/mayan/Downloads/5000-HRA-Records (2)/5000 HRA Records.csv")

plt.style.use('ggplot')  
plt.figure(figsize=(10, 6))
age_income = df.groupby('Age')['MonthlyIncome'].mean()
plt.plot(age_income.index, age_income.values, marker='o')
plt.title('Average Monthly Income by Age')
plt.xlabel('Age')
plt.ylabel('Average Monthly Income')
plt.grid(True)
plt.tight_layout()
plt.show()


gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff'])
plt.title('Gender Distribution')
plt.axis('equal')
plt.show()


plt.figure(figsize=(10, 6))
plt.hist(df['MonthlyIncome'], bins=30, color='teal', edgecolor='black')
plt.title('Monthly Income Distribution')
plt.xlabel('Monthly Income')
plt.ylabel('Frequency')
plt.show()



plt.figure(figsize=(10, 6))
years_income = df.groupby('YearsAtCompany')['MonthlyIncome'].mean()
plt.fill_between(years_income.index, years_income.values, color='skyblue', alpha=0.7)
plt.plot(years_income.index, years_income.values, color='blue')
plt.title('Average Monthly Income by Years at Company')
plt.xlabel('Years at Company')
plt.ylabel('Average Monthly Income')
plt.grid(True)
plt.tight_layout()
plt.show()



heatmap_data = df.groupby('MaritalStatus').agg({
    'MonthlyIncome': 'mean',
    'JobSatisfaction': 'mean',
    'Attrition': lambda x: (x == 'Yes').mean() * 100  # Attrition rate as %
})
heatmap_data.columns = ['Avg Monthly Income', 'Avg Job Satisfaction', 'Attrition Rate (%)']
data_matrix = heatmap_data.values
labels = heatmap_data.columns
marital_labels = heatmap_data.index
plt.figure(figsize=(8, 5))
plt.imshow(data_matrix, cmap='YlGnBu', aspect='auto')
plt.colorbar(label='Value')
plt.xticks(ticks=np.arange(len(labels)), labels=labels, rotation=45)
plt.yticks(ticks=np.arange(len(marital_labels)), labels=marital_labels)
for i in range(data_matrix.shape[0]):
    for j in range(data_matrix.shape[1]):
        val = round(data_matrix[i, j], 1)
        plt.text(j, i, f'{val}', ha='center', va='center', color='black')
plt.title('Heatmap: Marital Status vs Key Metrics')
plt.tight_layout()
plt.show()



```
